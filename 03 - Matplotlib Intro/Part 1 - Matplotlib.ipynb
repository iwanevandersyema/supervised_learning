```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,10,100)
y = np.sin(X)
```

## MATLAB Style

```python
plt.plot(X,y)
```

## Matplotlib Style

```python
fig = plt.figure() #optional
ax = plt.axes()
ax.plot(X,y)
```

# Standard example

## MATLAB Style

```python
plt.figure(figsize=(6, 4))
plt.plot(X,y, "r--", linewidth=1, label="sin(X)")
plt.plot(X, np.cos(X), "bo", markersize=3, label="cos(X)")
plt.xlim(0,14)
plt.ylim(-2,2)
plt.title("Fungsi Trigonometri")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(); # supaya tidak keluar output baris terkahir, kasih ";" atau disimpan di variable kosong _ = plt.legend();
```

## Matplotlib Style

```python
fig = plt.figure(figsize=(6, 4))
ax = plt.axes()
ax.plot(X,y, "r--", linewidth=1, label="sin(X)")
ax.plot(X, np.cos(X), "bo", markersize=3, label="cos(X)")
ax.set(xlim=(0,14), ylim=(-2,2), title="Fungsi Trigonometri", xlabel="x", ylabel="y")
ax.legend();
```

# Subplots

## MATLAB Style

```python
plt.subplot(2,3,1) #bisa pake 231
plt.plot(X,y, "g*")
plt.plot(X,np.cos(X), "co" )

plt.subplot(2,3,3) #bisa pake 233
plt.plot(X,y, "r--")

plt.subplot(2,3,5) #bisa pake 235
plt.plot(X,np.cos(X), "co" )
```

## Matplotlib Style

```python
fig,ax = plt.subplots(2,3, figsize=(8,6))
ax[0,0].plot(X,y)
ax[0,2].plot(X,np.cos(X))
ax[1,1].plot(X,np.cos(X), "co" )
```
