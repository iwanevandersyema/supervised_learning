```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/linear.csv")
df.head()
```

```python
df_1 = df[df.tipe == "data1"]
df_2 = df[df.tipe == "data2"]

plt.scatter(df_1.X, df_1.y)
plt.scatter(df_2.X, df_2.y)
```

```python
sns.scatterplot("X", "y", data=df, hue="tipe")
```

```python
plt.figure(figsize=(8,8))
sns.lmplot("X", "y", data=df, hue="tipe")
```

```python
sns.distplot(df.y)
```

# Coba seaborn di data cereal

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

## Pairplot

```python
sns.pairplot(df, vars=["calories", "fiber", "rating"])
```

## Categorical Data (X = kategorik, y=numerik)

## Persebaran

```python
sns.catplot("mfr", "rating", data=df)
```

```python
sns.boxplot("mfr", "rating", data=df)
```

```python
sns.violinplot("mfr", "rating", data=df)
```

## Frekuensi

```python
sns.countplot("mfr", data = df, hue="shelf")
```
